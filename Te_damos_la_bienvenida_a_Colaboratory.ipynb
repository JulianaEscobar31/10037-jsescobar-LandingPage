```
# Importar la biblioteca necesaria
from transformers import pipeline

# Cargar el pipeline para preguntas y respuestas
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# Definir el contexto en el que se buscará la respuesta
context = """
Hugging Face es una empresa que desarrolla herramientas y modelos para el procesamiento del lenguaje natural (NLP).
Sus productos más conocidos son los modelos Transformers, que han revolucionado el campo de NLP.
Hugging Face también proporciona una plataforma para el desarrollo y la implementación de modelos de machine learning.
"""

# Definir la pregunta sobre el contexto
question = "¿Qué productos son los más conocidos de Hugging Face?"

# Usar el pipeline para encontrar la respuesta
result = qa_pipeline(question=question, context=context)

# Imprimir el resultado
print(result)
```
